```python
import numpy as np

# Define the factor nodes (factor potentials) as Gaussian distributions.
def factor_potential(x, f):
    return np.exp(-0.5 * (x - f)**2)

# Define the number of values each variable node can take
num_values_per_node = 10

# Initialize variable nodes with random values between 0 and 1
variable_nodes = np.random.uniform(0, 1, (3, num_values_per_node))

# Initialize feature nodes
feature_nodes = [0.5, 0.5]  # Initialize feature nodes

# Initialize messages from variables to features and from features to variables
messages_v2f = [np.ones((num_values_per_node, 2)) for _ in range(2)]  # Messages from variables to features
messages_f2v = [np.ones((num_values_per_node, 3)) for _ in range(3)]  # Messages from features to variables

# Perform belief propagation iterations (assuming convergence)
num_iterations = 50

for iteration in range(num_iterations):
    # Update messages from variables to features
    for f in range(2):
        for value_index in range(num_values_per_node):
            for v in range(3):
                if v != f:
                    message = 1.0
                    for vf in range(2):
                        if vf != f:
                            message *= messages_f2v[v][value_index, vf]
                    messages_v2f[f][value_index, v] = message

    # Update messages from features to variables
    for v in range(3):
        for value_index in range(num_values_per_node):
            for f in range(2):
                message = 1.0
                for vf in range(2):
                    if vf != v:
                        message *= messages_v2f[vf][value_index, f]
                message *= factor_potential(variable_nodes[v][value_index], feature_nodes[f])
                messages_f2v[v][value_index, f] = message

# Compute the marginal probabilities for each value of each variable node
marginal_probabilities = [messages_f2v[v] / np.sum(messages_f2v[v], axis=1, keepdims=True) for v in range(3)]

# Print the marginal probabilities for each variable node
for v in range(3):
    print(f"Marginal probabilities for Variable Node {v}:")
    for value_index, probabilities in enumerate(marginal_probabilities[v]):
        print(f"Value {value_index}: {probabilities}")
```

```python
import numpy as np
from scipy.stats import norm

# Define variables
num_values = 5  # Number of possible values each variable can take
num_iterations = 10  # Number of iterations for belief propagation

# Initialize messages from variable nodes to factor nodes and vice versa
var_to_factor = {
    'x1f1': np.ones(num_values),
    'x2f1': np.ones(num_values),
    'x2f2': np.ones(num_values),
    'x3f2': np.ones(num_values)
}

factor_to_var = {
    'f1x2': np.ones(num_values),
    'f2x2': np.ones(num_values)
}

# Potential tables for factor nodes (replace with actual potential tables)
# For simplification, using Gaussian distribution
f1_potential = norm.pdf(np.arange(num_values), loc=2, scale=1)
f2_potential = norm.pdf(np.arange(num_values), loc=3, scale=1)

# Belief propagation algorithm
for _ in range(num_iterations):
    # Update messages from variable nodes to factor nodes
    var_to_factor['x1f1'] = np.ones(num_values)  # x1 has only one neighbor, f1
    var_to_factor['x2f1'] = factor_to_var['f2x2']
    var_to_factor['x3f2'] = np.ones(num_values)  # x3 has only one neighbor, f2
    var_to_factor['x2f2'] = factor_to_var['f1x2']

    # Update messages from factor nodes to variable nodes
    # Multiply incoming messages with potential and sum out all variables except the target variable
    factor_to_var['f1x2'] = f1_potential * var_to_factor['x1f1']
    factor_to_var['f1x2'] = np.sum(factor_to_var['f1x2'])  # Sum out other variables (if any)
    factor_to_var['f2x2'] = f2_potential * var_to_factor['x3f2']
    factor_to_var['f2x2'] = np.sum(factor_to_var['f2x2'])  # Sum out other variables (if any)

# Compute final beliefs for each variable
belief_x1 = var_to_factor['x1f1']  # Since x1 has only one neighbor
belief_x2 = factor_to_var['f1x2'] * factor_to_var['f2x2']
belief_x3 = var_to_factor['x3f2']  # Since x3 has only one neighbor

# Normalize beliefs
belief_x1 /= np.sum(belief_x1)
belief_x2 /= np.sum(belief_x2)
belief_x3 /= np.sum(belief_x3)

# Print most probable values
print('Most probable value for x1:', np.argmax(belief_x1))
print('Most probable value for x2:', np.argmax(belief_x2))
print('Most probable value for x3:', np.argmax(belief_x3))
```
